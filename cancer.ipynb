```python
X=df.iloc[:,2:32].values
Y=df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)

```

```python
x_test
```

```python
#function for model creation
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
lin.fit(x_train,y_train)


from sklearn.linear_model import LogisticRegression
log=LogisticRegression()


log.fit(x_train,y_train)

#deciosn Tree
    
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(criterion='entropy', random_state=0)
tree.fit(x_train,y_train)

#randomforrest

from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
forest.fit(x_train,y_train)
print("Linear Regression Accuracy",lin.score(x_train,y_train))

print("LogisticRegression Accuracy",log.score(x_train,y_train))
print("DecisionTreeClassifier Accuracy",tree.score(x_train,y_train))
print("RandomForestClassifier Accuracy",forest.score(x_train,y_train))



```
